Analyze the distribution of numerical variables

```python


# Select numerical columns for analysis
numerical_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

# Descriptive statistics
print(data[numerical_cols].describe())

# Histograms
data[numerical_cols].hist(bins=20, figsize=(12, 6))
plt.suptitle('Histograms of Numerical Variables', fontsize=14)
plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[numerical_cols])
plt.title('Box Plots of Numerical Variables', fontsize=14)
plt.ylabel('Value')
plt.show()

# Pairplot
color_palette = sns.color_palette("Set2")
sns.pairplot(data[numerical_cols],palette=color_palette)
plt.suptitle('Pairplot of Numerical Variables', fontsize=14)
plt.show()

# Correlation matrix heatmap
correlation_matrix = data[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap', fontsize=14)
plt.show()
```

From the above graphs, several insights can be derived:
1. From the histogram;
The sales are primarily centeralized in a small range less than $5000. profit is the same with a negative range, indicating possible losses.
The Quantity of products most frequently purchased are 2,4, and 6 decreasing as the quantity increases. The Discount is rarely applied, and mostly none is applied followed by a discount of 20%.
2. From the boxplot;
Boxplot helps identify outliers, and clearly here, Sales and Profit have outliers.
3. The pairplot and the heatmaps identify possible correlation between the variables at hand. Positive correlation specifies that as one variable increase the other increases, and negative correlation specifies as one increases the other decreases. Sales and Profit decreases as discount increases.




Analyze and Visualize Categorical variables

```python


# Explore the Ship Mode column
ship_mode_counts = data['Ship Mode'].value_counts()
print(ship_mode_counts)

# Visualize the Ship Mode distribution
sns.countplot(x='Ship Mode', data=data)
plt.title('Distribution of Ship Modes')
plt.show()

# Explore the Segment column
segment_counts = data['Segment'].value_counts()
print(segment_counts)

# Visualize the Segment distribution
sns.countplot(x='Segment', data=data)
plt.title('Distribution of Customer Segments')
plt.show()

# Explore the Region column
region_counts = data['Region'].value_counts()
print(region_counts)

# Visualize the Region distribution
sns.countplot(x='Region', data=data)
plt.title('Distribution of Regions')
plt.show()

# Explore the State column
region_counts = data['State'].value_counts()
print(region_counts)


# Explore the Category column
category_counts = data['Category'].value_counts()
print(category_counts)

# Visualize the Category distribution
sns.countplot(x='Category', data=data)
plt.title('Distribution of Product Categories')
plt.show()

# Explore the Sub-Category column
sub_category_counts = data['Sub-Category'].value_counts()
print(sub_category_counts)

# Visualize the Sub-Category distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='Sub-Category', data=data)
plt.title('Distribution of Product Sub-Categories')
plt.xticks(rotation=45)
plt.show()
```

In-depth Exploratory Data Analysis

```python


# Explore sales by Region
sales_by_region = data.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(sales_by_region)

# Visualize sales by Region
plt.figure(figsize=(12, 6))
sales_by_region.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()

# Explore sales by Category
sales_by_Category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(sales_by_Category)

# Visualize sales by Category
plt.figure(figsize=(12, 6))
sales_by_Category.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()


# Explore sales by Quantity
sales_by_Quantity = data.groupby('Quantity')['Sales'].sum().sort_values(ascending=False)
print(sales_by_Quantity)

# Visualize sales by Quantity
plt.figure(figsize=(12, 6))
sales_by_Quantity.plot(kind='bar')
plt.title('Sales by Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.show()

# Explore sales by Discount
sales_by_Discount = data.groupby('Discount')['Sales'].sum().sort_values(ascending=False)
print(sales_by_Discount)

# Visualize sales by Discount
plt.figure(figsize=(12, 6))
sales_by_Discount.plot(kind='bar')
plt.title('Sales by Discount')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()

# Analyze profit margins by Region
profit_by_Region= data.groupby('Region')['Profit'].mean().sort_values(ascending=False)
print(profit_by_Region)

# Visualize profit margins by Region
plt.figure(figsize=(10, 6))
profit_by_Region.plot(kind='bar')
plt.title('Profit Margins by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()



# Analyze profit margins by Segment
profit_by_Segment = data.groupby('Segment')['Profit'].mean().sort_values(ascending=False)
print(profit_by_Segment)

# Visualize profit margins by Segment
plt.figure(figsize=(10, 6))
profit_by_Segment.plot(kind='bar')
plt.title('Profit Margins by Segment')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.show()


# Analyze profit margins by Discount
profit_by_Discount = data.groupby('Discount')['Profit'].mean().sort_values(ascending=False)
print(profit_by_Discount)

# Visualize profit margins by Discount
plt.figure(figsize=(10, 6))
profit_by_Discount.plot(kind='bar')
plt.title('Profit Margins by Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python

# Segment and Category Analysis
segment_sales = data.groupby('Segment')['Sales'].sum()
category_profit = data.groupby('Category')['Profit'].sum()

# Geographic Analysis
region_sales = data.groupby('Region')['Sales'].sum()
city_profit = data.groupby('City')['Profit'].sum()

# Product Analysis
subcat_sales = data.groupby('Sub-Category')['Sales'].sum()
product_profit = data.groupby('Sub-Category')['Profit'].sum()

# Discount Analysis
discount_sales = data.groupby('Discount')['Sales'].sum()
discount_profit = data.groupby('Discount')['Profit'].sum()

# Shipping Analysis
shipmode_sales = data.groupby('Ship Mode')['Sales'].sum()
shipmode_profit = data.groupby('Ship Mode')['Profit'].sum()



# Plotting the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
segment_sales.plot(kind='bar')
plt.title('Segment-wise Sales')

plt.subplot(2, 3, 2)
category_profit.plot(kind='bar')
plt.title('Category-wise Profit')

plt.subplot(2, 3, 3)
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')

plt.subplot(2, 3, 4)
city_profit.nlargest(10).plot(kind='bar')
plt.title('Top 10 Cities by Profit')

plt.subplot(2, 3, 5)
subcat_sales.plot(kind='bar')
plt.title('Sub-Category Sales')

plt.subplot(2, 3, 6)
product_profit.plot(kind='bar')
plt.title('Sub-Category Profit')

plt.tight_layout()
plt.show()

print('Discount vs. Sales:')
print(discount_sales)
print('\nDiscount vs. Profit:')
print(discount_profit)

print('\nShipping Mode vs. Sales:')
print(shipmode_sales)
print('\nShipping Mode vs. Profit:')
print(shipmode_profit)


```
